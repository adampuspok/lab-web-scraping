```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Base URL
base_url = 'https://books.toscrape.com/catalogue/page-{}.html'

# Empty lists for all data BEFORE the loop
book_titles = []
book_prices = []
book_stars = []
availabilities = []

# Loop through 50 pages
for page in range(1, 51):
    print(f"Scraping page {page}...")
    url = base_url.format(page)
    response = requests.get(url)
    response.encoding = 'utf-8'  # Force correct encoding
    soup = BeautifulSoup(response.text, 'html.parser')

    components = soup.find_all('article', class_='product_pod')

    # Loop through each book on the page
    for component in components:
        # Title
        title = component.h3.a['title']
        book_titles.append(title)

        # Price
        price = component.find('p', class_='price_color').get_text()
        price = price.replace('Â', '')  # Remove encoding artifact if any
        price_num = float(price.replace('£', '').strip())
        book_prices.append(price_num)

        # Rating
        rating = component.find('p', class_='star-rating')['class'][1]
        book_stars.append(rating)

        # Availability
        availability = component.find('p', class_='instock availability').get_text(strip=True)
        availabilities.append(availability)

# Create DataFrame after all pages processed
df = pd.DataFrame({
    'Book Title': book_titles,
    'Price': book_prices,
    'Rating': book_stars,
    'Availability': availabilities
})

print(df.shape)  # Should print (1000, 4)
print(df.head())
```

```python
#Now I need the additional book infos, that are stored on separate pages per books (UPC, Genre, Description):
#BASIC CODE:
import requests
from bs4 import BeautifulSoup
import pandas as pd

def scrape_main_info():
    base_url = 'https://books.toscrape.com/catalogue/page-{}.html'
    titles = []
    prices = []
    ratings = []
    availabilities = []

    detail_urls = [] 

    for page in range(1, 3):  # adjust to 51 for all pages
        print(f"Scraping main page {page}...")
        url = base_url.format(page)
        response = requests.get(url)
        soup = BeautifulSoup(response.text, 'html.parser')

        books = soup.find_all('article', class_='product_pod')
        for book in books:
            titles.append(book.h3.a['title'])
            
            price_text = book.find('p', class_='price_color').get_text()
            price_text = price_text.replace('Â', '')
            price = float(price_text.replace('£', '').strip())
            prices.append(price)

            rating_str = book.find('p', class_='star-rating')['class'][1]
            ratings.append(rating_str)

            availability = book.find('p', class_='instock availability').get_text(strip=True)
            availabilities.append(availability)

            detail_url = book.h3.a['href'].replace('../../../', '')
            full_detail_url = 'https://books.toscrape.com/catalogue/' + detail_url
            detail_urls.append(full_detail_url)

    df_main = pd.DataFrame({
        'Title': titles,
        'Price (£)': prices,
        'Rating': ratings,
        'Availability': availabilities
    })
    return df_main, detail_urls

def get_detail_info(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')

    upc = soup.find('th', string='UPC').find_next_sibling('td').text
    genre = soup.find('ul', class_='breadcrumb').find_all('li')[2].a.text

    desc_tag = soup.find('div', id='product_description')
    description = desc_tag.find_next_sibling('p').text if desc_tag else ''

    return upc, genre, description


df, urls = scrape_main_info()

upcs = []
genres = []
descriptions = []

for url in urls:
    print(f"Fetching details for {url}")
    upc, genre, description = get_detail_info(url)
    upcs.append(upc)
    genres.append(genre)
    descriptions.append(description)

# Add detailed info columns to df
df['UPC'] = upcs
df['Genre'] = genres
df['Description'] = descriptions

print(df.head())
```

```python
# Convert star rating to numeric
rating_map = {
    'One': 1,
    'Two': 2,
    'Three': 3,
    'Four': 4,
    'Five': 5
}


df['Rating_num'] = df['Rating'].map(rating_map)

# Ask user for input
min_rating = float(input("Enter minimum rating (e.g., 4): "))
max_price = float(input("Enter maximum price (e.g., 20): "))

filtered_df = df[
    (df['Rating_num'] >= min_rating) &
    (df['Price (£)'] <= max_price)
].drop(columns='Rating_num')  # Drop helper column if not needed

# Show result
print(f"\nFiltered books with rating >= {min_rating} and price <= £{max_price}:")
print(filtered_df.head())
```
